### **Perceptron Learning**

Here, we'll use a Graph library to create a  network. This isn't exaclt what we do in Machine Learinig, but it's a great way to create visualization that you can modify in code

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import math
```

Let's build a simple Perceptron using the library `networkx`

It'll consist of three basic nodes
- Inputs
- Activations (conections between all nodes)
- Outputs 


```python
dense = nx.Graph()

# defining the layers
inputs = {i: (0, i) for i in range(0, 5)}
activations = {i+100: (1, i) for i in range(0, 5)}
outputs = {i+1000: (2, i) for i in range(0, 5)}

all = {**inputs, **activations, **outputs}
```

And now, we'll fully connect every input so that it can talk with all the activations.

```python
for input in inputs:
  for activation in activations:
    dense.add_edge(input, activation)

for activation in activations:
  for output in outputs:
    dense.add_edge(activation, output)
```

```python
nx.draw_networkx_nodes(dense, all, nodelist=all.keys(), node_color='b')
nx.draw_networkx_edges(dense, all, edge_color='r')

axes = plt.axis('off')
```

You can see, how our Neural Network looks like. 

But in practice, these graphs are represented as tensors at each layer and are connected via operations, such as a tensor product which mathematically connects nodes via multiplication and addition
